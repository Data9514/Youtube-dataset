```python
#3.Check for the Percentage of the missing values and drop or impute them.
from sklearn.impute import SimpleImputer


missing_percentage = df.isnull().mean() * 100
print("Missing Percentage:")
print(missing_percentage[missing_percentage > 0])
threshold = 50  
columns_to_drop = missing_percentage[missing_percentage > threshold].index
df = df.drop(columns=columns_to_drop)
imputer = SimpleImputer(strategy='mean')
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

print("After Handling Missing Values:")
```

```python
# 4.Check the statistical summary of both numerical and categorical columns and write your inferences.
df.describe()
df.describe(include ='O')
```

```python
# 5. Convert datatype of column published_at from object to pandas datetime.
df['published_at'] = pd.to_datetime(df['published_at'])
df.dtypes
```

```python
# 6.Create a new column as 'published_month' using the column published_at (display the months only)
df['published_month'] = df['published_at'].dt.month
print(df[['published_at', 'published_month']])
```

```python
# 7.Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb' and so on..... 
df['published_month'] = df['published_at'].dt.month.map({1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'})
print(df[['published_at', 'published_month']])
```

```python
# 8. Find the number of videos published each month and arrange the months in a decreasing order based on the video count.
# Group by 'published_month' and count the number of videos
monthly_video_count = df['published_month'].value_counts().reset_index()
monthly_video_count.columns = ['Month', 'Video Count']

# Arrange months in decreasing order based on video count
monthly_video_count = monthly_video_count.sort_values(by='Video Count', ascending=False)


print(monthly_video_count)
```

```python
# 9. Find the count of unique video_id, channel_id and channel_title.
unique_video_id_count = df['video_id'].nunique()
unique_channel_id_count = df['channel_id'].nunique()
unique_channel_title_count = df['channel_title'].nunique()

# results
print("Unique Video ID Count:", unique_video_id_count)
print("Unique Channel ID Count:", unique_channel_id_count)
print("Unique Channel Title Count:", unique_channel_title_count)
```

```python
# 10.Find the top10 channel names having the highest number of videos in the dataset and the bottom10 having lowest number of videos.

channel_video_counts = df['channel_title'].value_counts()

# Top 10 channels with the highest number of videos
top_10_channels = channel_video_counts.head(10)

# Bottom 10 channels with the lowest number of videos
bottom_10_channels = channel_video_counts.tail(10)

#  results
print("Top 10 Channels (Highest Number of Videos):")
print(top_10_channels)

print("\nBottom 10 Channels (Lowest Number of Videos):")
print(bottom_10_channels)
```

```python
# 11.Find the title of the video which has the maximum number of likes and the title of the video having minimum likes and write your inferences.

max_likes_video_title = df.loc[df['likes'].idxmax()]['title']

# Find the title of the video with the minimum number of likes
min_likes_video_title = df.loc[df['likes'].idxmin()]['title']

#  results
print("Video with the Maximum Likes:")
print("Title:", max_likes_video_title)
print("Number of Likes:", df['likes'].max())

print("\nVideo with the Minimum Likes:")
print("Title:", min_likes_video_title)
print("Number of Likes:", df['likes'].min())
```

```python
# 12.Find the title of the video which has the maximum number of dislikes and the title of the video having minimum dislikes and write your inferences.

max_dislikes_video_title = df.loc[df['dislikes'].idxmax()]['title']

# Find the title of the video with the minimum number of dislikes
min_dislikes_video_title = df.loc[df['dislikes'].idxmin()]['title']

# Display the results
print("Video with the Maximum Dislikes:")
print("Title:", max_dislikes_video_title)
print("Number of Dislikes:", df['dislikes'].max())

print("\nVideo with the Minimum Dislikes:")
print("Title:", min_dislikes_video_title)
print("Number of Dislikes:", df['dislikes'].min())
```

```python
# 13.Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.

df['view_count'] = pd.to_numeric(df['view_count'], errors='coerce')
df['dislikes'] = pd.to_numeric(df['dislikes'], errors='coerce')
correlation_coefficient = df['view_count'].corr(df['dislikes'])
plt.figure(figsize=(10, 6))
plt.scatter(df['view_count'], df['dislikes'], alpha=0.5)
plt.title('Scatter Plot of Views vs Dislikes')
plt.xlabel('Number of Views')
plt.ylabel('Number of Dislikes')
plt.show()
print(f"Correlation Coefficient between Views and Dislikes: {correlation_coefficient}")
```

```python
# 14. Display all the information about the videos that were published in January, and mention the count of videos that were published in January.

df['published_at'] = pd.to_datetime(df['published_at'])

# Create a new column 'published_month' with the names of the months
df['published_month'] = df['published_at'].dt.month.map({1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'})

# Filter the DataFrame for videos published in January
january_videos = df[df['published_month'] == 'Jan']

# Display information about videos published in January
print("Information about Videos Published in January:")
print(january_videos)

# Mention the count of videos published in January
january_video_count = january_videos.shape[0]
print(f"\nCount of Videos Published in January: {january_video_count}")
```
